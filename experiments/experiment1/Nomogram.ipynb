```r
source("https://raw.githubusercontent.com/achekroud/nomogrammer/master/nomogrammer.r") 
library(ggplot2)
library(scales)
p <- nomogrammer(Prevalence = 0.25, Plr = 7.1, Nlr = 1.0, NullLine = TRUE, Detail = TRUE) +
     scale_color_manual(values=c("#FFFFFF","#417EFF")) + 
     theme(panel.grid.major = element_line(colour = "lightgrey")) +
     labs(colour = "") +
     theme(legend.position = "none")
# plot(p)
ptotal = p + ggtitle("Chest Pain Radiation", subtitle = "Both left and right arm")
ptotal
ggsave(ptotal, file = "nomogram.png",
       width = 5,
       height = 5, units = "in")
```
